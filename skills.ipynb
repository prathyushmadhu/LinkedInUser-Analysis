```python
# To remove fake usernames in linkedIn
def remove_dup_username(s):
    length = len(s)
    for i in range(length):
        if(s[i] == '&' or s[i] == '%' or s[i] == '^' or s[i] == '#' or s[i] == '@' or s[i] == '!'):
            return True
        elif(length >= 20):
            return True
    return False
# print(dm[0][0]) 

# print(remove_dup_username(dm[0][0]))
i=0
while(i<len(dm)):
    if(remove_dup_username(dm[i][0])):
        dm = np.delete(dm,i, axis=0)
    else:
        i+=1
print(dm) 
```

### Most important skillset for software engineer

```python
# Grouping Professionals who are working as software engineers/developers.
def identify_software(dm_1):
    dm_2 = []
    for i in range(len(dm_1)):
        if ('software' in dm_1[i][1]):
            dm_2.append(dm_1[i])
    return dm_2 

"""for i in range(len(dm_1)):
        for j in range(len(dm_1[i][1])):
            if ('software' in (dm_1[i][1][j].split(" "))):
                print(dm_1[i])   """

# for i in range(len(dm_1)):
#for j in range(len(dm_1[362][1])):
"""print(dm_1[362][1])
print(len(dm_1[362][1]))
if 'software' in dm_1[362][1]:
    print("true")
else:
    print("false") """

dm_2 = identify_software(dm_1)
print(dm_2)
```

```python
len(dm_2)
```

```python
# Placing all the skills of software developers/engineer into a list
def skillset_ordering(dm_2):
    skills_list = []
    for i in range(len(dm_2)):
        dm_2[i][2] = dm_2[i][2].replace('[','')
        dm_2[i][2] = dm_2[i][2].replace(']','')
        list = dm_2[i][2].split(',')
        for skills in list:
            skills_list.append(skills)
    return skills_list
    
skills_list = skillset_ordering(dm_2)
print(skills_list)
```

```python
# Gives the count of skills
counts = pd.Series(skills_list).value_counts()
counts
```

OBSERVATION : We can infer that the most popular skills required for the role of a software developer/software engineer are C++, java, javascript, sql and mysql.

### Job that higly soughts python 

```python
# Grouping professionals who knows python.
def identify_python(dm_1):
    dm_3 = []
    for i in range(len(dm_1)):
        if ('python' in dm_1[i][2]):
            dm_3.append(dm_1[i])
    return dm_3 

dm_3 = identify_python(dm_1)
print(dm_3)
```

```python
len(dm_3)
```

```python
# Inserting jobs of those professional into a list.
def job_ordering(dm_3):
    jobs_list = []
    for i in range(len(dm_3)):
        dm_3[i][1] = dm_3[i][1].replace('[','')
        dm_3[i][1] = dm_3[i][1].replace(']','')
        list = dm_3[i][1].split(',')
        for jobs in list:
            jobs_list.append(jobs)
    return jobs_list
    
jobs_list = job_ordering(dm_3)
print(jobs_list)
```

```python
# Gives the count of jobs
counts = pd.Series(jobs_list).value_counts()
counts
```

### popular skills for a professional who knows react js

```python
# Grouping professionals who knows react js.
def identify_react(dm_1):
    dm_4 = []
    for i in range(len(dm_1)):
        if  ('react' in dm_1[i][2]):
            dm_4.append(dm_1[i])
    return dm_4 

dm_4 = identify_react(dm_1)
print(dm_4)
```

```python
len(dm_4)
```

```python
# Listing other skills of the above professionals.
def react_skills(dm_4):
    skills_list = []
    for i in range(len(dm_4)):
        dm_4[i][2] = dm_4[i][2].replace('[','')
        dm_4[i][2] = dm_4[i][2].replace(']','')
        list = dm_4[i][2].split(',')
        for skills in list:
            skills_list.append(skills)
    return skills_list
    
react_skills = react_skills(dm_4)
print(react_skills)
```

```python
counts = pd.Series(react_skills).value_counts()
counts
```

OBSERVATION : As we are looking for popular skills of professionals who knows React js, its obvious that React js tops the list.
1) The popular skill set that professionals (who knows react) have are javascript, node.js, html and css
